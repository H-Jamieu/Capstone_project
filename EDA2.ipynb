```python
def find_unique(inlist):
    o_list = []
    for i in inlist:
        if i not in o_list:
            o_list.append(i)
    return o_list
def retrive_unique(pddata):
    large_list = []
    for row in pddata:
        large_list += row
    uniqueSet = set(large_list)
    return list(uniqueSet)
```

```python
def create_one_hot(df,ref_list,sourceRow):
    for r in ref_list:
        df[r] = 0
    for i in range(0,len(df)):
        for r in ref_list:
            if r in df.loc[i,sourceRow]:
                df.at[i,r] = 1
```

```python
def generateUnique_sig(df, col):
    cand = []
    for row in df[col]:
        cand.append(row)
    unique = set(cand)
    return list(unique)
def create_one_hot_sig(df, ref, col):
    for r in ref:
        df[r] = 0
    for idx, row in df.iterrows():
        for r in ref:
            if row[col] == r:
                df.at[idx,r] = 1
```

```python
series = generateUnique_sig(df_metadata, 'series')
original_language = generateUnique_sig(df_metadata, 'original_language')
```

```python
consums = df_metadata.loc[:,df_metadata.columns[-(160):]].sum()
colsums = df_metadata.loc[:,df_metadata.columns[-(160+1695):-160]].sum()
langsums = df_metadata.loc[:,df_metadata.columns[-(160+1695+89):-(160+1695)]].sum()
print(consums.index)
print(colsums.index)
print(langsums.index)
```

```python
def dropLeastColumns(ref, tr):
    dropList = []
    for idx, cnt in ref.iteritems():
        if cnt<tr:
            dropList.append(idx)
    return dropList
```

```python
conlist = dropLeastColumns(consums, 11)
langlist = dropLeastColumns(langsums, 100000)
collist = dropLeastColumns(colsums,3)
print(len(conlist))
print(len(langlist))
print(len(collist))
```
